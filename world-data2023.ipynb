```python
comma_cols=[]
percent_cols=[]
dollar_cols=[]
for col in df.columns:
    if df[col].dtype=="object":
        if df[col].str.contains(",").any():
            comma_cols.append(col)
        elif df[col].str.contains("%").any():
            percent_cols.append(col)
        elif df[col].str.contains("$").any():
            dollar_cols.append(col)

```

```python
for col in percent_cols:
    df[col]=df[col].str.replace("%","")
```

```python
for col in dollar_cols:
    df[col]=df[col].str.replace("$","",regex=True)
```

```python
for col in comma_cols:
    df[col]=df[col].str.replace(",","")
```

```python

    df["GDP"]=df["GDP"].str.replace("$","",regex=True)
```

```python
for col in df.columns:
    if df[col].dtype=="object":
        if col == 'Density\n(P/Km2)'or col =='Land Area(Km2)'or col =='Armed Forces size'or col =='Co2-Emissions'or col =='CPI'or col =='GDP'or col =='Population'or col =='Urban_population'or col =='Agricultural Land( %)'or col =='CPI Change (%)'or col =='Forested Area (%)'or col =='Gross primary education enrollment (%)'or col =='Gross tertiary education enrollment (%)'or col =='Out of pocket health expenditure'or col =='Population: Labor force participation (%)'or col =='Tax revenue (%)'or col =='Total tax rate' or col =='Unemployment rate' or col =='Gasoline Price' or col =='Minimum wage':
            df[col]=df[col].astype("float64")
```

# <span style="color:#090448"> Top ten high and low birth rate Countries  </span>
 we will analyse and visualise **high and  low birth rate countries**.

```python
birth_rate =df[["Country","Birth Rate"]].dropna().sort_values(by="Birth Rate",ascending=False).reset_index()
```

```python
highest_birth_rate_countries=birth_rate[["Country","Birth Rate"]].head(10).reset_index().drop("index",axis=1)
```

```python
highest_birth_rate_countries
```

```python
lowest_birth_rate_countries=birth_rate[["Country","Birth Rate"]].tail(10).reset_index().drop("index",axis=1)
```

```python
lowest_birth_rate_countries
```

```python
ax=highest_birth_rate_countries.sort_index(ascending=False).set_index("Country").plot(kind="barh",color="#843C0C",legend=False)
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0],color="blue",fontsize=7)

plt.title("top ten countries with highest birth rate",color="#2A1304")
plt.xlabel("birth rate",color="#2A1304")
plt.ylabel("Country",color="#2A1304")

ax=lowest_birth_rate_countries.sort_index(ascending=False).set_index("Country").plot(kind="barh",color="#C55A11",legend=False)
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0],color="blue",fontsize=7)

plt.title("top ten countries with lowest birth rate",color="#843C0C")
plt.xlabel("birth rate",color="#843C0C")
plt.ylabel("Country",color="#843C0C")
```
