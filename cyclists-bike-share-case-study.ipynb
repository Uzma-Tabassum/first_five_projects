## <span style="color:#3F95A4"> Combining datasets </span>

Now, I will combine all datasets in one data Frame using **concat** method.

```python
data1 = pd.concat([may,june,july,august],ignore_index=True, sort =False)
```

```python
data2 = pd.concat([september,october,november,december],ignore_index=True, sort =False)
```

```python
data3 = pd.concat([january,febuary,march,april],ignore_index=True, sort =False)
```

```python
data = pd.concat([data1,data2,data3],ignore_index=True, sort =False)
```

## <span style="color:#3F95A4"> Data Cleaning and Formatting</span>
First of all, I opened all csv files in **Excel**. I looked all files one by one and I found

```python
# checking data type
data.dtypes
```

```python
# formatting date columns into datetime
data["started_at"]=pd.to_datetime(data["started_at"],format="%Y-%m-%d %H:%M:%S")
data["ended_at"]=pd.to_datetime(data["ended_at"],format="%Y-%m-%d %H:%M:%S")
```

```python
# checking data type again
data.dtypes
```

as we can see, date columns have been changed into datetime.

```python
# deleting unnecessary columns
data =data.drop(labels=['start_station_name','start_station_id','end_station_name','end_station_id','start_lat','start_lng','end_lat','end_lng'],axis=1)
```

## <span style="color:#3F95A4">Data Analysis </span>

I am analyzing data. I will analyse the following,
- how casual riders and annual members use bikes .
- finding ride length for casual rides and annual member.
- total rides of casual riders and annual members.
- total rides in each month.
- common days of weekdays. 
etc.


```python
# making new column for ride length
data["ride_length"]=data["ended_at"] - data["started_at"]
```

```python
# making new column for weekdays
data["week_days"]=data["started_at"].dt.day_name()
```

```python
# making new column for month
data["month"]=data["started_at"].dt.month_name()
```

```python
# describing data
data.describe(include="all", datetime_is_numeric=True)
```

as we can see, Saturday is most common and busy day. July is most common month. average ride length is 18 minutes and 56 seconds. electric bike is favourite ride. annual members are greater than casual riders
